```python
import cv2
```

```python
img_cv=cv2.imread('2018-12-18-16-40-18-116.jpg')
type(img_cv)
```

```python
img_cv.shape
```

```python
plt.imshow(img_cv)  # o/ looks differnet cuz the order of color channel is different in open Cv(BGR) PIL(RGB)
```

```python
# change the color space with conversion code

new_cv_img=cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
plt.imshow(new_cv_img)
```

```python
# use slices to obtain differnet color channel
#Slicing selects each row and column.The final index selects the color channel.We can view the color channels blue, green and red. 

blue,green,red=img_cv[:,:,0],img_cv[:,:,1],img_cv[:,:,2] # assign each channel to corresponding array blue green and red

```

```python
plt.imshow(green)
```
